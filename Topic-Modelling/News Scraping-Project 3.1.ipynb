## News scraping from different fields

```python
#importing required libraries
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
#getting the url
url="https://timesofindia.indiatimes.com/business/india-business/deep-contraction-in-gdp-on-expected-line-recovery-to-be-gradual-says-industry/articleshow/77857692.cms"
res=requests.get(url)
res
```

```python
#creating instance for beautiful soup scrapping text data
soup = BeautifulSoup(res.text, "html.parser")
```

```python
#removing the unwanted characters and tags from the data
article=" "
for i in soup.find_all('div', class_='_1_Akb clearfix'):
    article = article + ' ' +  i.text
print(article)
```

```python
#saving the data to local file
with open("D:/Fliprobo/NLP class notebooks/New folder/news1.txt", "w", encoding="utf-8") as f:
    f.write(article)
```

### Scraping 2nd news data regarding jee-neet exams

```python
#getting the url
url2="https://timesofindia.indiatimes.com/home/education/entrance-exams/jee-neet-2020-vmc-launches-rapid-success-courses-for-entrance-exam-preparations/articleshow/77365635.cms"
res2=requests.get(url2)
res2
```

```python
#instance creation for text data using beautiful soup
soup2 = BeautifulSoup(res2.text, "html.parser")
soup2
```

```python
#extracting the data using tag id/class
content2 = soup2.find('div', {"class": "section1"})
```

```python
#printing the data extracted
content2
```

```python
#Creating the text data by removing unwanted tags and ids
article2 = ''
for i in content2.findAll('div'):
    article2 = article2 + ' ' +  i.text
print(article2)
```

```python
#appending the data to existing text file
with open('D:/Fliprobo/NLP class notebooks/New folder/news1.txt', 'a') as file:
    file.write(article2)
```
